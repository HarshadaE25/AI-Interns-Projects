##What is the distributions Females and males across the dataset?

```python
import pandas as pd
import matplotlib.pyplot as plt

file_name = 'Students_Data.csv'

# 1. Load the Data
df = pd.read_csv(file_name)

# 2. Calculate Gender Counts
gender_counts = df['Gender'].value_counts()
print("\n--- Gender Counts ---")
print(gender_counts)

# 3. Create the Modern Donut Chart
plt.style.use('ggplot')

plt.figure(figsize=(8, 8))

# Define parameters for the Donut Chart
labels = gender_counts.index
sizes = gender_counts.values
colors = ['#ff9999', '#66b3ff']  # Pastel colors (Female/Male)
explode = [0.03, 0.03]  # Slightly separating the slices

plt.pie(
    sizes,
    labels=labels,
    autopct='%.1f%%',  # Shows percentage with one decimal place
    startangle=90,
    explode=explode,
    colors=colors,
    shadow=True,
    textprops={'fontsize': 14, 'color': 'black', 'fontweight': 'bold'}
)

# Create the circle in the middle to transform the pie into a donut
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Distribution of Male and Female Students', fontsize=16, fontweight='bold', color='#333333')
plt.axis('equal') # Ensures the chart is circular
plt.tight_layout()

# Save the plot
output_file = 'gender_distribution_donut_chart.png'
plt.savefig(output_file)
```

## What is the mean score of female and male in all the exams?

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Students_Data.csv")

# Group by 'Gender' and calculate the mean of the score columns
mean_scores_by_gender = df.groupby('Gender')[['Maths Score', 'Reading Score', 'Writing Score']].mean()

# Reset index to make 'Gender' a column for a cleaner output
mean_scores_by_gender = mean_scores_by_gender.reset_index()

# Display the resulting DataFrame
print(mean_scores_by_gender.to_markdown(index=False, numalign="left", stralign="left"))
```

## Which particular Ethnicity group has the highest mean score in the reading Exam?

```python
import pandas as pd

# Assuming 'df' is the DataFrame loaded from 'Students_Data.csv'
# If starting fresh, you would load the file first:
# df = pd.read_csv("Students_Data.csv")

# Group by 'Ethnicity' and calculate the mean of 'Reading Score'
mean_reading_score_by_ethnicity = df.groupby('Ethnicity')['Reading Score'].mean()

# Convert the Series to a DataFrame and reset the index
mean_reading_score_by_ethnicity = mean_reading_score_by_ethnicity.reset_index()

# Rename the column for clarity
mean_reading_score_by_ethnicity.columns = ['Ethnicity', 'Mean Reading Score']

# Sort the DataFrame by 'Mean Reading Score' in descending order
mean_reading_score_by_ethnicity_sorted = mean_reading_score_by_ethnicity.sort_values(
    by='Mean Reading Score',
    ascending=False
)

# Display the resulting DataFrame
print(mean_reading_score_by_ethnicity_sorted.to_markdown(index=False, numalign="left", stralign="left"))
# Removed the undefined variable call: result_df
```

# The ethnicity group with the highest mean score in the Reading Exam is Group E with a mean score of approximately $73.03$.

```python
import pandas as pd
import plotly.express as px

# Data provided by the user in the previous step
data = {
    'Ethnicity': ['group E', 'group D', 'group C', 'group B', 'group A'],
    'Mean Reading Score': [73.02, 70.03, 69.10, 67.35, 64.67]
}

df_plot = pd.DataFrame(data)

# Ensure the data is sorted by score in descending order for the visualization
df_plot = df_plot.sort_values(by='Mean Reading Score', ascending=False)

# Create an interactive bar chart using Plotly Express
fig = px.bar(
    df_plot,
    x='Ethnicity',
    y='Mean Reading Score',
    title='Mean Reading Score by Ethnicity Group',
    labels={'Mean Reading Score': 'Mean Reading Score', 'Ethnicity': 'Ethnicity Group'},
    color='Mean Reading Score',  # Color the bars based on the score
    color_continuous_scale=px.colors.sequential.Viridis,
    text='Mean Reading Score' # Display the score on top of the bars
)

# Customize the text format to two decimal places
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

# Ensure the y-axis starts at a reasonable value or 0 for a bar chart
fig.update_yaxes(range=[0, df_plot['Mean Reading Score'].max() + 5])

# Save the plot as a JSON file for the interactive display
fig.write_json('mean_reading_score_by_ethnicity_bar_chart.json')

 # ---display the figure in colob---
fig.show()
```
