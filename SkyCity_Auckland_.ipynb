```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("SkyCity Auckland Restaurants & Bars.csv")

# -------------------------------------------------
# Step 1: Calculate Total Share
# -------------------------------------------------
df["Calculated_Total_Share"] = (
    df["InStoreShare"] +
    df["UE_share"] +
    df["DD_share"] +
    df["SD_share"]
)

# -------------------------------------------------
# Step 2: Define Tolerance (for rounding error)
# -------------------------------------------------
tolerance = 1  # Allow ±1%

# -------------------------------------------------
# Step 3: Identify Inconsistent Rows
# -------------------------------------------------
df["Share_Difference"] = df["Calculated_Total_Share"] - 100

inconsistent_rows = df[
    df["Share_Difference"].abs() > tolerance
]

# -------------------------------------------------
# Step 4: Print Results
# -------------------------------------------------
print("Total Restaurants:", len(df))
print("Number of Share Inconsistencies:", len(inconsistent_rows))

# Show problematic records (if any)
inconsistent_rows[[
    "RestaurantID",
    "RestaurantName",
    "Calculated_Total_Share",
    "Share_Difference"
]]
```
